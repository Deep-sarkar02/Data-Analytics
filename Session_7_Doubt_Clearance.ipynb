#Q-3 --Compare spending in "Electronics" vs. "Clothing"


```python
# we will get this in the items colm
# so get the unique items from the colm
data['Item'].unique()
```

```python
# we have to compare the both the spending
# we will have to apply the grroupby
# so we will seperate the electronics and the cloting
data['Item'].isin(['Electronics' , 'Cloths']) # this will give the boolean values
# this will give the data of the table which only have that items
# now  from the boolean we will convert it to the true data
data[data['Item'].isin(['Electronics' , 'Clothes'])]

# now we will groupby it on basis of the item
data[data['Item'].isin(['Electronics' , 'Clothes'])].groupby('Item')
# now we will find the total spent on basis of the items and then find the avg total spent
data[data['Item'].isin(['Electronics' , 'Clothes'])].groupby('Item')['Total Spent'].mean()
```

#What percentage of customers have made purchases in at least 5 different categories?


```python
# we will find the groupby on basis of the customers and then find the category
data.groupby('Customer ID')
# on basis of the customer id we will find the category
# nunique so we will find the number
cat_count= data.groupby('Customer ID')['Category'].nunique()
#this will give the uniwue category in whcih shopping have been done by the cutomer

```

```python
# we will now find where the cat count > 5
cat_count[cat_count>5]
# and we need the count
cat_count[cat_count>5].count()

# find the precentage
cat_count[cat_count>5].count()/ data['Customer ID'].nunique()
# so the 100% of the data
```
