#3D SURFACE PLOT

```python
# it can be used where value of z will be dependent on x , y
# and there must be 3 variable

# 3 component x , y , z
# it will print all the output of z for every combination of x , y

# length , breadth and area -- > area

# for lenght
# generate  99 diff valus from  0 to 100
l = np.linspace(10 , 100 , 99 )
# for breadth
b = np.linspace(5 , 70 , 99)

# now we want to know that for this possible values of length and breadth , how the area will be chnaging
# we willl use meshgrid function
# we will find the all the posssible combinations for l and b
import numpy as np
x , y = np.meshgrid(l , b)

# now we can do the any arithmetic op over this 2 variable
# find z or area or , x ,y
#z = x  * y


# also do another eqn
z = x**2 + y**2


# now we will create the surface plot for the l , b , z
fig = plt.figure(figsize = (15 ,15))
ax = plt.subplot(projection= "3d")

# now plot
p = ax.plot_surface(x , y , z  ,cmap = "viridis")

# add the color bar
fig.colorbar(p)
```

#and sometimes the client dont understand the 3d plot so we need to draw the contour plots

```python
fig = plt.figure(figsize = (5 , 5))
ax = plt.subplot()

# now use  ax.contour
p = ax.contour(l , b  , z , cmap ='viridis' )
fig.colorbar(p)

```

#contourfill - contourf

```python
# we can properly study the plot by the contourf
fig = plt.figure(figsize = (5 , 5))
ax = plt.subplot()

# now use  ax.contour
p = ax.contourf(l , b  , z , cmap ='viridis' )
fig.colorbar(p)

```

#GRIDSPEC

```python
from matplotlib.gridspec import GridSpec as g
```

```python
x = np.random.randn(100) # random 100 values
y = np.random.randn(100) # 100 values

# fig space
fig = plt.figure(figsize = (10 ,8))
# call the g
# now we will split the fig in 3 , 3
gs = g(3,3 , figure = fig)

print(gs)









# now we can add the subplot over this
# 1st subplot will occupy the space from oth col to all
ax1 = fig.add_subplot(gs[0 , 0 : ])# on the 0 th row
# plot data in ax1
ax1.plot(x ,y)


# for the ax2
ax2 = fig.add_subplot(gs[1 ,  : -1])
ax2.plot(x , y )

#for ax3
ax2 = fig.add_subplot(gs[1 ,  -1: ])
ax2.scatter(x , y )


```
