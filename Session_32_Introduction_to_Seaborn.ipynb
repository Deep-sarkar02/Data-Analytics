```python
# import the details
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

#**why we need seaborn ??**

```python
data = pd.read_csv("/content/spending_patterns_detailed.csv")
data

# Pivot table
abc = pd.pivot_table(data, values='Total Spent', index='Location', columns='Payment Method', aggfunc='sum').reset_index()

# Plotting
plt.figure(figsize=(20, 6))
x = np.arange(abc["Location"].nunique())  # X-axis base positions

# Bars with offset
plt.bar(x - 0.2, abc["Cash"], width=0.1, label="Cash")
plt.bar(x - 0.1, abc["Credit Card"], width=0.1, label="Credit Card")
plt.bar(x, abc["Debit Card"], width=0.1, label="Debit Card")
plt.bar(x + 0.1, abc["Digital Wallet"], width=0.1, label="Digital Wallet")

# Labels and legend
plt.xticks([0.0, 1.0, 2.0], ["Instore", "Mobile App", "Online"])
plt.legend()
plt.show()
```

#**LINEPLOT**

```python
# use the gapminder datset
import plotly.express as px
gap = px.data.gapminder()
gap
```

```python
#LINEPLOT OVER THIS DATA
# for the le.
sns.lineplot(data = gap ,
             x= "year",
             y = "lifeExp",

             # hue param --. differenciate by the the hue
             hue = "continent",

             # style param on basis of the style
             style = "continent",


             #we can reduce the confidance interval  range
             # by default it will be 95
             #so reset it to 95
             #ci --> confidence interval
             errorbar = ("ci" , 95))


# so at each year  it willl find the mean of the data of the y axis and plot that data
# and the shaded area is the confidance interval of  the lifeexpectancey

```

```python
# but if we dont want the mean  of the data we want any other opearion
# so we can use the param estimator

sns.lineplot(data = gap ,
             x= "year",
             y = "lifeExp",

             # hue param --. differenciate by the the hue
             hue = "continent",

             errorbar = ("ci" , 0),
             estimator = "max") # so the max values
```

#**Histogram (axis level)**

```python
# devide the data in bins
# tips
tips
```

```python
# plot
sns.histplot(data = tips ,
             x= "total_bill",
             # devide on basis of the sex
             hue = "sex",
             # what will be the no. of bins we want
             bins = 10,

             # now in the y axis we need to counts the density of the data
             stat = "density",
             # now we can define the range of the bin
             binrange = (10 , 60),

             # now if we want to see the cumulative freq
             cumulative = False ,


             # now we can decide the style of the bar
             # chnage from stacked to the side by side
             multiple = "dodge",

             # plot the distribution
             # kde
             kde = True, # how  the data pointsa re arrenged

             #now we can change the pallete
             palette = "Set2" #  or the style of the graph
             )
```

#**KDE PLOT**

```python
# shows the density and the distribution
sns.kdeplot(data =tips ,
            x = "total_bill",
            # diff on based of the sex
            hue = "sex" ,

            # in the disstribution  it will be filled by color
            fill = True,

            #now we want to make it stack
            multiple = "stack", # default
            # change the palletee
            palette = "Set2",


            # now we can define the range the number
            # in the x axis
            clip = (-10 , 70)


            )
```
