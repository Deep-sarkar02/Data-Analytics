#Identify customers whose monthly spending remains consistent. Define consistency as having a coefficient of variation (standard deviation/mean) of monthly spending below 10%.


```python
dr=df.groupby(['Customer ID' , 'year-month'])['Total Spent'].sum().reset_index()
```

```python
# now on the basis of the cust id , find the mean and the std
dr2=dr.groupby('Customer ID')['Total Spent'].agg(["mean" , "std"]).reset_index()
```

```python
# d=find the std / mean
dr2['cov'] = (dr2["std"] / dr2["mean"]) *100
```

```python
# we need only less than 10
dr2[dr2['cov'] < 10]
```
