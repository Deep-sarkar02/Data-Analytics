```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import plotly.express as px
```

#working with plotly go

```python
# figure

# create 2 data where we can plot the graph
x = [1 , 2,3, 4,5]
y = [10 , 15 , 17,19 ,21]

# create figure
# wiht go
fig = go.Figure()



# now creta the chart
# now we will add the chart to the figure
#fig.add_trace()

# now we will add trhe scatter plot inside that
#fig.add_trace(go.Scatter(x = x , y  = y))
#fig.add_trace(go.Scatter(x = x , y  = y , mode = "lines"  ,name = 'Sample line'))
fig.add_trace(go.Scatter(x = x , y  = y , mode = "lines+markers"  ,name = 'Sample line'))







#now we can add the title and other by the help of the layout
fig.update_layout(
    title = "plotly graph",
    xaxis_title = "x axis",
    yaxis_title = "y axis"
)
```

```python
 # another way
 # efirst create the trace and layout
 # while creating figure , pass the trace and layout
trace  = go.Scatter(x = x , y  = y , mode = "lines+markers"  ,name = 'Sample line')

# layout
layput = go.Layout(title = "plotly graph",
    xaxis_title = "x axis",
    yaxis_title = "y axis")


# create the figure and pass the both
fig = go.Figure(trace , layput)

```

#updattion on the chart

```python
#now chnage the shape of the marker
# now we will use the update trace
t.update_traces(marker =dict(symbol = "diamond") )
```

```python
#change the color and the size of the marker
t.update_traces(marker =dict(symbol = "cross" ,
                             color = "red",
                             size = 5) )
```

#updation on the layout

```python
# update layout


# update the bg color of the plotting area
t.update_layout(plot_bgcolor = "yellow")

# chnage the bg color of the outer backgroud
t.update_layout(paper_bgcolor = "pink")


# change the font of the overall and dssign it
t.update_layout(font = dict(family = "Arail" ,
                            size = 10,
                            color = 'red',
                            weight = 'bold'))
```

```python
# we can also add the title
t.update_layout(title = dict(
                              #test of the title
                              text = "this is the title",
                              #font of the title
                              font = dict(size = 20 ,
                                          color = "black",
                                          weight = "bold"),
                              # we want the title from the center of the x ie , 0.5
                              x = 0.5
))
```

```python
# also chnage in the x axis
t.update_layout(xaxis =
                dict(
                    title = "hdi",
                    titlefont = dict(
                                      size = 14,
                                      color = "blue"
                                      ),
                    # now if we want to chnage the style of the numbers
                    tickfont = dict(
                                      size = 14,
                                      color = "green"
                                      ),
                    # chnage the color of the grid in the xaxis
                    gridcolor = "brown"
                    ))
```

```python
#same for y axis
t.update_layout(yaxis =
                dict(
                    title = "hdi",
                    titlefont = dict(
                                      size = 14,
                                      color = "blue"
                                      ),
                    # now if we want to chnage the style of the numbers
                    tickfont = dict(
                                      size = 14,
                                      color = "green"
                                      ),
                    # chnage the color of the grid in the xaxis
                    gridcolor = "brown"
                    ))
```
