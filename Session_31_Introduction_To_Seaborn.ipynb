```python
import numpy as np
import plotly.graph_objects as go

# Create data
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2

# Create 3D surface plot
trace = go.Contour(x=x, y=y, z=z)
layout = go.Layout(
    title="3D Surface Plot",
    height=900,
    width=800
)

go.Figure(trace, layout)

```

```python
import seaborn as sns

# Load the tips dataset
tips = sns.load_dataset("tips")
```

```python
#  create a scatter plot for the
# bill vs tips for each of the gender
# now normally we nedd to make 2 scatter plot one for the female and other for the male

# by the facet plot we can make it  easily
import plotly.express as px

px.scatter(tips , x = 'total_bill' , y = 'tip' , facet_col='sex')

# so on the same plotting area is devided in to 2 diff category\

# and facet_col means  the plotting will be on the based of the colm
# there is another called facet_row where the plotting willl be done on the based of the row
```

```python
# on the based of row
px.scatter(tips , x = 'total_bill' , y = 'tip' , facet_row='sex')

```

```python
# both facet roe and the col
#on the category :-  sex and the smoker
px.scatter(tips , x = 'total_bill' , y = 'tip' , facet_row='sex' , facet_col = "smoker")
```

```python
# change the color on the basis of the time
px.scatter(tips , x = 'total_bill' , y = 'tip' , facet_row='sex' , facet_col = "smoker" ,color = 'time')
```

#**SUBPLOTS BY PLOTLY**

```python
# we can make sublpots by the new module
# called make_subplots


# import the detsils
from plotly.subplots import make_subplots
# define the figure
# and define the col and the row of the subplots
fig = make_subplots(rows = 2 ,
                    cols = 2  # grid of 2 and 2
                    ,subplot_titles=('A' ,'B' , 'C' , 'D')
                    )
fig
```

```python
# now we can define that what will be type of the plots we can
# define in each of the plotiig area
# by the help of the param spec
fig = make_subplots(rows = 2 ,
                    cols = 2  # grid of 2 and 2
                    ,subplot_titles=('A' ,'B' , 'C' , 'D'),
                    specs = [[{"type" : "xy"} , {"type" : "xy"}] ,  # 1st row xy means the 2 axis plots ie ,scatter plot,line plot , and it is also  by default
                              [{"type" : "xy"} , {"type" : "domain"}] # domain for the piechart
                            ]

                    )
#for 3d plot use scene
#  for polar graph use ploar
# for map base grpah use mapbox
# for domain use the pechart , sunburst , treemap
```

```python
# now add the  line plot for the row and col 1
fig.add_trace(go.Scatter(x = [1 , 2,3,4,5] ,
                         y = [10 ,15 , 17 , 19 , 20] , mode = "lines+markers" , name= "Line plot for the cell 0" , line = dict(color = "blue") ) , row = 1 ,col=1)
# for the 1 row and the 2 col
# crerate a bar chart
fig.add_trace(go.Bar(x = ['A' , 'B' , 'C' ,'D'] ,
                     y = [5 , 10 , 15 , 20 ] , name = "barchatrt for the 2 nd cell" , marker = dict(color = "orange")) , row = 1 , col=2)


# sctter plot for the r2 col1
fig.add_trace(go.Scatter(x = [1, 2,3,4,5,] ,
                     y = [5 , 10 , 15 , 20 , 45  ] ,mode = "markers",  name = "sctter for the 3 rd cell") , row = 2 , col=1)


# for the nextr it will be a piechart
fig.add_trace(go.Pie(labels  = ["Red" , "Blue" , "Green"] ,
                     values = [5 ,14 , 23 ] , name = "piechart for the 4 th  cell" , showlegend=False ) , row = 2 , col=2)



# update the height and the other caharecters of the subplot
fig.update_layout(title_text = "subplot in plotly" ,
                  height = 600,
                  width = 800 ,
                  # show the legend
                  showlegend = True,
                  # update the legend\
                  legend = dict(x= 1.25 ,
                                bgcolor = "pink" ,
                                bordercolor = "blue"),
                  #chnage the bg color of the plotting area
                  plot_bgcolor = "lightgreen",
                   # also chnage the bgcolor of  the full area
                  paper_bgcolor = "cyan"
                  )
```
