#SCATTER PLOT

```python
# WE will use the seaborn inbuilt dataset
import seaborn as sns
sns.get_dataset_names()
# here we will use the tips dataset
```

```python
data= sns.load_dataset('tips')
```

```python
data.head()
```

```python
# now we will plot the scatter plot by the plt.scatter() formula
plt.scatter(data['total_bill'] ,data['tip'] )
```

```python
# change the shape of the market
plt.scatter(data['total_bill'] ,data['tip'], marker = '*' )

# chnage the size of the marker
plt.scatter(data['total_bill'] ,data['tip'], marker = "*" , s=20 )


#chnage the color of the marker
plt.scatter(data['total_bill'] ,data['tip'], marker = "*" , s=20  ,color = 'red')


# now we can also chnge the boundary color of the market
plt.scatter(data['total_bill'] ,data['tip'], marker = "*" , s=20  ,color = 'red' , edgecolors='blue')



# now we can aslo chnage the transparencey of the marker
# by the param called alpha
plt.scatter(data['total_bill'] ,data['tip'], marker = "*" , s=20  ,color = 'red' , edgecolors='blue' , alpha = 0.1)




# now we can add the grid
plt.grid()


# we can add the title and xlabel , y label and all the other fegtures from the line plot
plt.title("total bill vs tips")



# xlabel
plt.xlabel('total bill')

# ylabel
plt.ylabel("tips")

#
```

```python
# plot total sales as per data
abc = data.groupby('Date')[['Sales' , 'Profit' , 'Quantity']].sum().reset_index()
```

```python
abc
```

```python
# add the quantity colm
abc['Quantity'] = [10 ,13 , 25 ,67, 100 , 63, 14 ,14 , 26 ,34]
```

```python
abc
```

```python
# now we will pplot the scatter plot
plt.scatter(abc['Sales'] , abc['Profit'])

# label
plt.xlabel('Sales')
plt.ylabel("profit")


# now if we want to chanage the marker size as per the quantity
plt.scatter(abc['Sales'] , abc['Profit'] , s= abc['Quantity'])



# the above is called multivariate analysis since we are using more than 2 colms
plt.grid()
```
